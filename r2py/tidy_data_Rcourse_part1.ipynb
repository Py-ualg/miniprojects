```python
# Sidenote:
def get_var_name(var) -> str:
    """ This can be used for python variables, not DFs """
    for name, value in globals().items():
        if value is var:
            return name
```

```python
# instead we define manually name attribute
# this is experimental pandas functionality but if you save DFs as pickle
# it gets saved and can be loaded back
rf_s_q1.attrs['name'] = "rf_s_q1"
rf_w_q1.attrs['name'] = "rf_w_q1"
rf_s_q2.attrs['name'] = "rf_s_q2"
rf_w_q2.attrs['name'] = "rf_w_q2"

ra_s_q1.attrs['name'] = "ra_s_q1"
ra_w_q1.attrs['name'] = "ra_w_q1"
ra_s_q2.attrs['name'] = "ra_s_q2"
ra_w_q2.attrs['name'] = "ra_w_q2"
```

```python
# now attributes hash table
col_vals_dict = {
    'rf': 'Ria Formosa',
    'ra': 'Ria de Alvor',
    's': 'summer',
    'w': 'winter',
    'q1': 'q1',
    'q2': 'q2',
}
```

```python
# iterable of datasets
dataset_list = [
    rf_s_q1, rf_w_q1, rf_s_q2, rf_w_q2,
    ra_s_q1, ra_w_q1, ra_s_q2, ra_w_q2,
]
```

```python
# FINALLY, iterate over the list, pass certain part of the name to the colcol_vals_dict
# So, next time, you run only this modulo name changes.
for ds in dataset_list:
    rf_s_q1['region'] = "Ria Formosa"
    ds['region'] = col_vals_dict[ds.attrs['name'][:2]]
    ds['season'] = col_vals_dict[ds.attrs['name'][3:4]]
    ds['quadrat'] = col_vals_dict[ds.attrs['name'][5:7]]
    
```

```python
# crabs07 <-
#   crabs06 %>%
#   dplyr::mutate(
#     region = factor(region),
#     season = factor(season),
#     quadrat = as.integer(quadrat),
#     species = factor(species),
#     sex = factor(sex, levels = c("female", "male")),
#     stage = factor(stage, levels = c("juvenile", "pre_puberty", "sub_adult", "adult"), ordered = TRUE)
#   )
crabs07 = crabs06.copy()
crabs07[['region', 'season', 'species', 'sex', 'quadrat']] = crabs07[['region',
                                                                      'season',
                                                                      'species',
                                                                      'sex',
                                                                      'quadrat']].astype('category')

# for custom ordering
cat_dtype = CategoricalDtype(
    categories=["juvenile", "pre_puberty", "sub_adult", "adult"], ordered=True)
crabs07['stage'] = crabs07['stage'].astype(cat_dtype)
crabs07['quadrat'] = crabs07['quadrat'].cat.codes
```

```python
# quadrats02 <-
#   quadrats01 %>%
#   dplyr::select(-1) %>%
#   tidyr::pivot_longer(cols = everything(), names_to = "region_quadrat", values_to = "area") %>%
#   tidyr::separate(col = "region_quadrat", into = c("region", "quadrat"), sep = "-")

# Drop the first column
quadrats02 = quadrats01.iloc[:, 1:]

# Reshape the DataFrame from wide to long format
quadrats02 = quadrats02.melt(var_name='region_quadrat', value_name='area')

# Separate the 'region_quadrat' column into 'region' and 'quadrat'
quadrats02[['region', 'quadrat']] = quadrats02['region_quadrat'].str.split('-', expand=True)

# Drop the 'region_quadrat' column
quadrats02.drop(columns=['region_quadrat'], inplace=True)

# reorder columns
quadrats02 = quadrats02[['region', 'quadrat', 'area']]
```

```python
quadrats02
```

```python
quadrats03 = quadrats02.copy()
quadrats03[['region', 'quadrat']] = quadrats03[['region', 'quadrat']].astype('category')
quadrats03['quadrat'] = quadrats03['quadrat'].cat.codes
```

```python
# quadrats_coord <- dplyr::distinct(crabs07, region, quadrat, latitude, longitude)
# quadrats04 <- dplyr::left_join(quadrats03, quadrats_coord)

# # And now we can drop latitude and longitude from the `crabs07` data set.
# # Note the minus sign behind the vector of column names.
# crabs08 <- select(crabs07, -c("latitude", "longitude"))

# Create quadrats_coord DataFrame containing distinct values of region, quadrat, latitude, and longitude
quadrats_coord = crabs07[['region', 'quadrat', 'latitude', 'longitude']].drop_duplicates()

# Merge quadrats03 with quadrats_coord using left join
quadrats04 = pd.merge(quadrats03, quadrats_coord, on=['region', 'quadrat'], how='left')

# Drop latitude and longitude columns from crabs07
crabs08 = crabs07.drop(columns=['latitude', 'longitude'])
```
